```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, output_file, show


x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 25, 15]


p = figure(title="Simple Bokeh Plot", x_axis_label='X', y_axis_label='Y')

p.line(x, y, legend_label='Line', line_width=2)

output_file("bokeh_plot.html")
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
import pandas as pd

df = pd.read_csv('data.csv')

p = figure(title='Scatter plot with circles', x_axis_label='x', y_axis_label='y')

p.circle(x='x', y='y', size=10, color='blue', source=df)

output_file('scatter.html')
show(p)
```

```python
from bokeh.plotting import figure, output_file, show


x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 25, 15]


p = figure(title="Custom Axis Labels", x_axis_label='X Axis', y_axis_label='Y Axis')

p.line(x, y)


p.xaxis.major_label_orientation = 'vertical'
p.xaxis.major_label_text_font_size = '16pt'


p.yaxis.major_label_orientation = 'horizontal'
p.yaxis.major_label_text_font_size = '16pt'


output_file("custom_axis_labels.html")
show(p)
```

```python
from bokeh.plotting import figure, output_file, show


x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 25, 15]


p = figure(title="Custom Title", x_axis_label='X Axis', y_axis_label='Y Axis')


p.line(x, y)


p.title.align = 'center'
p.title.text_font_size = '24pt'


output_file("custom_title.html")
show(p)
```
